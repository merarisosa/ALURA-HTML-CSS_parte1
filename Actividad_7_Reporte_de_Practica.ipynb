## combinar DataFrame

```python
import pandas as pd

DF1 = pd.DataFrame({
    'NOMBRE': ['Jose', 'Max'],
    'CARRERA': ['Economia', 'Arquitectura'],
    'EDAD': [23, 26]
}).set_index('NOMBRE')

DF2 = pd.DataFrame({
    'NOMBRE': ['Aurora', 'Maria'],
    'CARRERA': ['Medicina', 'Informatica'],
    'EDAD': [22, 28]
}).set_index('NOMBRE')

DF = pd.concat([DF1, DF2])

DF
```

```python
DF1 = pd.DataFrame({
    'AUTOS': ['Nissan', 'Ford', 'Audi'],
    'COLOR': ['blanco', 'Azul', 'Rojo']
}).set_index('AUTOS')

DF2 = pd.DataFrame({
    'AUTOS': ['Nissan', 'Ford', 'Audi'],
    'MODELO': [2018, 2020, 2022]
}).set_index('AUTOS')

DF1
DF2
```

```python
DF = pd.concat([DF1, DF2], axis= 1)
DF
```

```python

DF1 = pd.DataFrame({
    'AUTOS': ['Nissan', 'Ford', 'Audi'],
    'COLOR': ['blanco', 'Azul', 'Rojo']
})

DF2 = pd.DataFrame({
    'AUTOS': ['Nissan', 'Ford', 'Audi'],
    'MODELO': [2018, 2020, 2022]
})

DF = pd.merge(DF1, DF2, on='AUTOS', how='outer')

DF
```
